```python
# lugar donde mas se hace click en el dispositivo
plt.figure(figsize=(20, 30))

plt.scatter(click['touchX'], click['touchY'], 
            c='green', label= 'click',  edgecolor= '#eae8e8')

plt.title('Touch en el dispositivo',size=30)
plt.xlabel('eje X', size=20)
plt.ylabel('eje Y',size=20)
plt.legend(loc='top_left')

```

```python
# analizando cuanto tiempo tarda un usuario de un dispositivo especifico, en hacer click
clickTimeToClickAndspecsBrand = click[['specs_brand','timeToClick']]
clickTimeToClickAndspecsBrand = clickTimeToClickAndspecsBrand.dropna()  ## elimina los nulos
clickTimeToClickAndspecsBrand.head()
```

```python
## Grafico
plt.figure(figsize=(5, 20) )
s=300

plt.scatter(clickTimeToClickAndspecsBrand['specs_brand'], clickTimeToClickAndspecsBrand['timeToClick'],  
            c='red',edgecolor= '#eae8e8', s=s)

plt.title('Tiempo en hacer click por usuario segun cada marca',size=25)
plt.xlabel('Tipo de Marca-Especifica', size=20)
plt.ylabel('Tiempo en milisegundos',size=20)
plt.legend(loc='top_left')
```
